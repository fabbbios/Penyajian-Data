```python
def convert_playtime_to_hours(playtime):
    playtime = str(playtime).lower().strip()
    if playtime == '1-5':  # Handle the specific '1-5' case by returning 3.0
        return 3.0
    elif 'jam' in playtime:
        try:
            # Handle cases like "1 jam", "2 jam", "1,5 jam"
            return float(playtime.replace(' jam', '').strip().replace(',', '.'))
        except ValueError:
            return None # Return None for unparsable "jam" entries
    elif 'menit' in playtime:
        try:
            # Handle cases like "30 menit", "40 menit"
            return float(playtime.replace(' menit', '').strip().replace(',', '.')) / 60
        except ValueError:
            return None # Return None for unparsable "menit" entries
    else:
        try:
            # Handle standalone numbers with comma or dot as decimal
            return float(playtime.replace(',', '.'))
        except ValueError:
            return None # Return None for any other unparsable entries


df_combined['Berapa jam rata-rata anda bermain roblox dalam sehari? (Ex. 1, 2, 3 jam)'] = df_combined['Berapa jam rata-rata anda bermain roblox dalam sehari? (Ex. 1, 2, 3 jam)'].apply(convert_playtime_to_hours)
display(df_combined['Berapa jam rata-rata anda bermain roblox dalam sehari? (Ex. 1, 2, 3 jam)'].describe())
```

```python
import pandas as pd

# Pastikan kolom sesuai dataset kamu
col = "Biasanya anda bermain roblox menggunakan device apa?"

# Normalisasi huruf biar seragam
df[col] = df[col].str.strip().str.lower()

# Mapping manual untuk kategori yang sama
mapping_device = {
    "hp": "mobile",
    "handphone": "mobile",
    "tablet": "mobile",
    "Tablet": "mobile",
    "phone": "mobile",
    "pc/laptop dan mobile": "mobile dan pc"
}

# Ganti nilai sesuai mapping
df[col] = df[col].replace(mapping_device)

# Untuk memastikan hasilnya
print(df[col].value_counts())
```

```python
import re

col = "Berapa jam rata-rata anda bermain roblox dalam sehari? (Ex. 1, 2, 3 jam)"

# Normalisasi huruf kecil dan spasi
df[col] = df[col].astype(str).str.lower().str.strip()

def normalize_jam(x):
    # Jika ada kata "menit" â†’ otomatis "Kurang dari 1 jam"
    if "menit" in x:
        return "Kurang dari 1 jam"

    # Tangani range, misalnya "2-3 jam"
    match_range = re.match(r'(\d+\.?\d*)\s*-\s*(\d+\.?\d*)', x)
    if match_range:
        a, b = map(float, match_range.groups())
        avg = (a+b)/2
        return "Kurang dari 1 jam" if avg < 1 else f"{int(avg) if avg.is_integer() else avg} jam"

    # Tangani angka tunggal (misalnya "1 jam", "0.5 jam", "sekitar 2")
    match_num = re.search(r'(\d+\.?\d*)', x)
    if match_num:
        val = float(match_num.group(1))
        if val < 1:
            return "Kurang dari 1 jam"
        return f"{int(val) if val.is_integer() else val} jam"

    return None  # kalau tidak cocok pola apapun

df[col + "_normalisasi"] = df[col].apply(normalize_jam)

# Cek hasil
print(df[[col, col + "_normalisasi"]].head(20))
```

# Diagram Polar

```python
# Variabel
col_faculty = "Fakultas"
col_friends = "Berapa jumlah teman yang anda miliki di roblox? (Ex. 202)"

# Hitung rata-rata jumlah teman per fakultas
avg_friends = df.groupby(col_faculty)[col_friends].mean()

categories = avg_friends.index.tolist()
values = avg_friends.values.tolist()

# Hitung sudut untuk tiap kategori
angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False).tolist()

# Tutup lingkaran dengan menambahkan titik pertama di akhir
values += values[:1]
angles += angles[:1]
categories += categories[:1]

# Plot radar chart
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))

ax.plot(angles, values, linewidth=2, linestyle='solid', color="orange")
ax.fill(angles, values, color="orange", alpha=0.25)

# Label kategori di sumbu
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories[:-1], fontsize=10)

ax.set_title("Diagram Polar: Rata-rata Jumlah Teman per Fakultas", fontsize=14, pad=20)

plt.show()
```

```python
# Variabel
col_hours = "Berapa jam rata-rata anda bermain roblox dalam sehari? (Ex. 1, 2, 3 jam)_normalisasi"

# Hitung frekuensi responden per kategori jam
counts = df[col_hours].value_counts().sort_index()

categories = counts.index.tolist()
values = counts.values.tolist()

# Sudut untuk tiap kategori
angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False).tolist()

# Tutup lingkaran dengan menambahkan titik pertama di akhir
values += values[:1]
angles += angles[:1]
categories += categories[:1]

# Plot radar / polar chart
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))

ax.plot(angles, values, linewidth=2, linestyle='solid', color="teal")
ax.fill(angles, values, color="teal", alpha=0.25)

# Label kategori di tiap sumbu
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories[:-1], fontsize=10)

ax.set_title("Diagram Polar: Distribusi Jam Bermain Roblox per Hari", fontsize=14, pad=20)

plt.show()
```
